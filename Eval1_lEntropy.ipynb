```python
import pandas as pd
import numpy as np

data = pd.read_csv('results.csv')

data = data.to_numpy()

data = data[:,1:]

data = [x[0] for x in data] 

data.sort()
```

```python
data
```

```python
dick = dict()
for element in data:
    if element not in dick:
        dick[element] = 1
    else:
        dick[element] += 1
values = list(dick.items())

import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = [7.00, 3.50]
plt.rcParams["figure.autolayout"] = True

ind = []
fre = []
for item in values:
   ind.append(item[0])
   fre.append(item[1])

plt.bar(ind, fre)

plt.show()
```

```python
print(np.mean(data))
print(np.std(data))
print(np.var(data))
print(np.median(data))
print(np.min(data))
print(np.max(data))
```
